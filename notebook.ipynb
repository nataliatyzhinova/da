6. Как присобачить порядковый номер к списку с повторяющимися элементами (соответсвенно, у повторяющихся строк должен быть один и тот же номер). Подсказка, это как в SQL, нужно сделать джойн с таблицей-справочником (предварительно создав ее). Ну или как угобнее.


```python
# Дан список с повторяющимися элементами
items = ["a", "a", "b", "c", "d", "d", "e", "b", "f", "a"]

# Создаем словарь
items_dict = {}

i = 1

for item in items:
    # Если элемент новый
    if item not in items_dict:
        items_dict[item] = i
        i += 1

print(items_dict)  

# Создаем список с порядковыми номерами
num = [items_dict[item] for item in items]
print(num)  
```

```python
# Можно сохранить начальный список и порядковые номера в таблице (датафрейме) для удобства

df = pd.DataFrame(items)
df['num'] = num
df
```

```python
# Можно отсортировать по порядковому номеру

df.sort_values(by=["num"])
```
